## Find the top 10 major project themes (using column 'mjtheme_namecode')
Column 'mjtheme_namecode' contains nested data. The nested data should be flattened in order to solve the exercise. This can be achieved by loading JSON file as string and then passing the column 'mjtheme_namecode' to json_normalize.

```python
# Load JSON file as a string
df1 = json.load((open('../Desktop/Springboard/data_wrangling_json/data_wrangling_json/data/world_bank_projects.json')))
```

```python
# Flatten 'mjtheme_namecode' column and save in df1
df1 = json_normalize(df1, 'mjtheme_namecode')
# Retrieve first 15 counts of df1
df1.head(15)
```

Observe that a few name values are missing.However value_count can be used to retrive the top 10 project themes using 'code'

```python
# Retrieve top 10 project themes
top_ten_proj_theme_count = df1.code.value_counts().head(10)
top_ten_proj_theme_count
```

## Create a dataframe with the missing names filled in
This can be achieved by sorting code and name columns, replacing missing names with NULL values, then replacing those by forwardfilling.

```python
# Load JSON file as a string
df1 = json.load((open('../Desktop/Springboard/data_wrangling_json/data_wrangling_json/data/world_bank_projects.json')))
# Flatten 'mjtheme_namecode' column and replace the missing values with Nan
df1 = json_normalize(df1, 'mjtheme_namecode').replace('', np.NaN)
df1.head(15)
```

```python
# Forwardfill Nan Values and display top 10 project theme names
df1 = df1.sort_values(by=['code','name']).fillna(method='ffill')
df1.name.value_counts().head(10)
```
